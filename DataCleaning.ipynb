```python
import os
import numpy as np
from PIL import Image
from google.colab import drive

drive.mount('/content/drive')
```

```python
directories =["/content/drive/MyDrive/GAIP_Project/ImageClassification/Raw/Cactus/","/content/drive/MyDrive/GAIP_Project/ImageClassification/Raw/NotCactus/"]
for directory in directories:
    count = 1
    for filename in os.listdir(directory):
        file = os.path.join(directory, filename)
        if os.path.isfile(file):
            try:
                im = Image.open(file)
                rgb_im = im.convert("RGB")
                if directory == directories[0]:
                    new_directory = "/content/drive/MyDrive/GAIP_Project/ImageClassification/Clean/Cactus/"
                else:
                    new_directory = "/content/drive/MyDrive/GAIP_Project/ImageClassification/Clean/NotCactus/"
                new_file = os.path.join(new_directory, f"{count}.JPEG")
                rgb_im.save(new_file)
                count += 1
            except:
                count += 1
                pass
```

```python
dir = "/content/drive/MyDrive/GAIP_Project/ImageClassification/Clean/Cactus/"
print(len([entry for entry in os.listdir(dir) if os.path.isfile(os.path.join(dir, entry))]),"Initial Cactus Dataset")
count = 0
data = np.genfromtxt("/content/drive/MyDrive/GAIP_Project/ImageClassification/redCactus.csv",delimiter=",")
for i in data:
    if os.path.exists(f"/content/drive/MyDrive/GAIP_Project/ImageClassification/Clean/Cactus/{int(i)}.JPEG"):
        os.remove(f"/content/drive/MyDrive/GAIP_Project/ImageClassification/Clean/Cactus/{int(i)}.JPEG")
        count += 1
print(count,"deleted from Cactus")
print(len([entry for entry in os.listdir(dir) if os.path.isfile(os.path.join(dir, entry))]),"Final Cactus Dataset")
```

```python
dir = "/content/drive/MyDrive/GAIP_Project/ImageClassification/Clean/NotCactus/"
print(len([entry for entry in os.listdir(dir) if os.path.isfile(os.path.join(dir, entry))]),"Initial NotCactus Dataset")
count = 0
data = np.genfromtxt("/content/drive/MyDrive/GAIP_Project/ImageClassification/redNotCactus.csv",delimiter=",")
for i in data:
    if os.path.exists(f"/content/drive/MyDrive/GAIP_Project/ImageClassification/Clean/NotCactus/{int(i)}.JPEG"):
        os.remove(f"/content/drive/MyDrive/GAIP_Project/ImageClassification/Clean/NotCactus/{int(i)}.JPEG")
        count += 1
print(count,"deleted from NotCactus")
print(len([entry for entry in os.listdir(dir) if os.path.isfile(os.path.join(dir, entry))])," Final NotCactus Dataset")
```
